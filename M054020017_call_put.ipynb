```python
import numpy as np
```

```python
k = 9100
premium_call = 179
premium_put = 185
interval = 1000
```

```python
ST = np.arange(k-interval,k+interval)
print  (ST)
```

```python
Payoff_longcall = np.maximum(ST-k,0) - premium_call
Payoff_shortcall = - Payoff_longcall
```

```python
import matplotlib.pyplot as plt
```

```python
# plot long and short call
plt.plot(ST, Payoff_longcall)
plt.plot(ST, Payoff_shortcall)
plt.show()
```

```python
Payoff_longput = np.minimum(ST-k,0) -  premium_put 
Payoff_shortput = - Payoff_longput
```

```python
# plot long and short put
plt.plot(ST, Payoff_longput)
plt.plot(ST, Payoff_shortput)
plt.show()
```

```python
k1 = 8500
k2 = 8700

# strategy : Bull spread
bull_longcall = np.maximum(ST-k1,0) - premium_call
bull_shortcall = premium_call - np.maximum(ST-k2,0)
bull_spread = bull_longcall + bull_shortcall

# plot bull spread with long and short call 
plt.plot(ST,bull_longcall)
plt.plot(ST, bull_shortcall)
plt.plot(ST,bull_spread)
plt.show()
```
